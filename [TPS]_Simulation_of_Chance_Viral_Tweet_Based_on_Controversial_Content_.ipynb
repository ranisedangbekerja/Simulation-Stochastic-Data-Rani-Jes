#Introduction

##**Purpose**

This simulation aims to understand the impact of controversial content on virality, considering various factors such as:
1. Number of Followers
2. Use of Hashtags and Mentions
3. Likes
4. Comments
5. Reposts

##**About the Dataset**

The dataset used in this simulation is derived from real data scraped from two Twitter accounts known for their controversial content. The focus is on how these variables interact to influence the virality of posts.

##**Definition of Controversial**

For the purpose of this study, controversial content is defined based on our subjective judgment. Specifically, we consider content related to LGBT topics as controversial. This definition is used consistently throughout the simulation.

# Get Data Accout A

This account declare that he is a enginerring girl but actually he is a engineering guy.

```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/content/DataAccountA.csv'
df = pd.read_csv(file_path)

# Print the entire DataFrame
print(df.to_string())
```

```python
# Data tweet dan retweet
tweets = [
    1, 0.5, 1, 0.5, 1, 1, 0.5, 0, 0.5, 1, 1, 0.5, 0, 0.5, 1, 1, 0, 0.5, 0, 0
]

retweets = [
    1, 1, 1, 1, 0, 1, 1, 1, 1
]

# Fungsi untuk menghitung probabilitas
def calculate_probabilities(data):
    total = len(data)
    controversial = sum(1 for d in data if d == 1)
    neutral = sum(1 for d in data if d == 0.5)
    not_controversial = sum(1 for d in data if d == 0)

    prob_controversial = controversial / total
    prob_neutral = neutral / total
    prob_not_controversial = not_controversial / total

    return {
        "controversial": prob_controversial,
        "neutral": prob_neutral,
        "not_controversial": prob_not_controversial
    }

# Menghitung probabilitas untuk tweet dan retweet
tweet_probabilities = calculate_probabilities(tweets)
retweet_probabilities = calculate_probabilities(retweets)

# Menampilkan hasil
print("Probabilitas Tweet:")
print(f"Kontroversial: {tweet_probabilities['controversial']:.3f}")
print(f"Netral: {tweet_probabilities['neutral']:.3f}")
print(f"Tidak Kontroversial: {tweet_probabilities['not_controversial']:.3f}")

print("\nProbabilitas Retweet:")
print(f"Kontroversial: {retweet_probabilities['controversial']:.3f}")
print(f"Netral: {retweet_probabilities['neutral']:.3f}")
print(f"Tidak Kontroversial: {retweet_probabilities['not_controversial']:.3f}")
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd

# Create DataFrames
data_A = {
    "Followers": [0]*29 + [1228],
    "Views": [1332, 0, 0, 0, 0, 0, 0, 794, 4400, 1900, 0, 562, 105, 377, 864, 0, 2800, 920, 396, 959, 502, 424, 0, 389, 302, 319, 0, 367, 270, 620],
}

data_B = {
    "Followers": [0]*29 + [2523],
    "Views": [1100, 1500, 2700, 2900, 1400, 7600, 1200, 1100, 0, 1200, 1100, 2700, 0, 6900, 0, 3300, 12000, 0, 1900, 6700, 0, 3500, 3500, 793, 707, 948, 873, 635, 3800, 0],
}

df_A = pd.DataFrame(data_A)
df_B = pd.DataFrame(data_B)

# Initialize figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Function to update the plot
def update(num):
    ax.clear()
    ax.plot(df_A.index[:num], df_A["Views"][:num], label="User A - Views", color="blue")
    ax.plot(df_B.index[:num], df_B["Views"][:num], label="User B - Views", color="green")
    ax.set_title("Views Over Time")
    ax.set_xlabel("Posts")
    ax.set_ylabel("Views")
    ax.legend()

# Create animation
ani = FuncAnimation(fig, update, frames=len(df_A), repeat=False)

# Save animation as GIF
ani.save("/content/twitter_views_animation.gif", writer="imagemagick", fps=2)

# Show plot
plt.close(fig)  # Close the figure to prevent it from displaying in this environment
"/content/twitter_views_animation.gif"
```
